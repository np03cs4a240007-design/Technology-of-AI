**Adding HDI Category Column**

```python
def hdi_category(hdi):
    if hdi < 0.550:
        return "Low"
    elif 0.550 <= hdi <= 0.699:
        return "Medium"
    elif 0.700 <= hdi <= 0.799:
        return "High"
    else:
        return "Very High"
```

**Outlier Detection**

```python
# HDI IQR
Q1_hdi = hdi_south_asia["hdi"].quantile(0.25)
Q3_hdi = hdi_south_asia["hdi"].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

lower_hdi = Q1_hdi - 1.5 * IQR_hdi
upper_hdi = Q3_hdi + 1.5 * IQR_hdi

# HDI Outliers
hdi_outliers = hdi_south_asia[(hdi_south_asia["hdi"] < lower_hdi) |
                               (hdi_south_asia["hdi"] > upper_hdi)]

hdi_outliers[["country", "hdi"]]
```

```python
# GNI per Capita IQR
Q1_gni = hdi_south_asia["gross_inc_percap"].quantile(0.25)
Q3_gni = hdi_south_asia["gross_inc_percap"].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

lower_gni = Q1_gni - 1.5 * IQR_gni
upper_gni = Q3_gni + 1.5 * IQR_gni

# GNI Outliers
gni_outliers = hdi_south_asia[(hdi_south_asia["gross_inc_percap"] < lower_gni) |
                               (hdi_south_asia["gross_inc_percap"] > upper_gni)]

gni_outliers[["country", "gross_inc_percap"]]
```

```python
# Combine HDI and GNI outliers
outliers = hdi_south_asia[
    (hdi_south_asia["hdi"] < lower_hdi) | (hdi_south_asia["hdi"] > upper_hdi) |
    (hdi_south_asia["gross_inc_percap"] < lower_gni) | (hdi_south_asia["gross_inc_percap"] > upper_gni)
]

outliers[["country", "hdi", "gross_inc_percap"]]
```

```python
#Scatter Plot Highlighting Outliers
plt.figure(figsize=(8,6))

# Normal countries
plt.scatter(
    hdi_south_asia["gross_inc_percap"],
    hdi_south_asia["hdi"],
    color='blue',
    label="Normal Countries"
)

# Outliers
plt.scatter(
    outliers["gross_inc_percap"],
    outliers["hdi"],
    color='red',
    label="Outliers"
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("GNI per Capita vs HDI (South Asia)")
plt.legend()
plt.show()
```

# Discussion

# - Countries highlighted as outliers have either:
  - **High GNI but moderate HDI** â†’ shows income concentration or weaker education/health.

  - **High HDI despite low GNI** â†’ effective social policies or human development investment.

# **HDI Disparity**

```python
#Compute HDI Range (Max â€“ Min)
south_asia_range = south_asia_df["hdi"].max() - south_asia_df["hdi"].min()
middle_east_range = middle_east_df["hdi"].max() - middle_east_df["hdi"].min()

print("South Asia HDI Range:", south_asia_range)
print("Middle East HDI Range:", middle_east_range)
```

```python
# Compute Mean and Standard Deviation of HDI
south_asia_mean = south_asia_df["hdi"].mean()
south_asia_std = south_asia_df["hdi"].std()

middle_east_mean = middle_east_df["hdi"].mean()
middle_east_std = middle_east_df["hdi"].std()
```

```python
# Compute Coefficient of Variation
south_asia_cv = south_asia_std / south_asia_mean
middle_east_cv = middle_east_std / middle_east_mean

print("South Asia HDI CV:", south_asia_cv)
print("Middle East HDI CV:", middle_east_cv)
```

# HDI Variation Conclusion

Based on the computed HDI range and coefficient of variation (CV):

- **South Asia exhibits more variation in HDI** than the Middle East.
- South Asia has a larger gap between its highest and lowest HDI values.
- The higher CV in South Asia indicates greater relative dispersion in human development levels.
- The Middle East shows more consistent HDI values across countries.

**Conclusion:**  
Human development is more uneven in South Asia, while the Middle East demonstrates lower HDI disparity.

# **Outlier Detection**

```python
# Function to Detect Outliers Using 1.5 Ã— IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] < lower) | (df[column] > upper)]
```

```python
# Detect Outliers in South Asia
sa_hdi_outliers = detect_outliers(south_asia_df, "hdi")
sa_gni_outliers = detect_outliers(south_asia_df, "gross_inc_percap")

print("South Asia HDI Outliers:")
print(sa_hdi_outliers[["country", "hdi"]])

print("\nSouth Asia GNI per Capita Outliers:")
print(sa_gni_outliers[["country", "gross_inc_percap"]])
```

```python
# Detect Outliers in Middle East
me_hdi_outliers = detect_outliers(middle_east_df, "hdi")
me_gni_outliers = detect_outliers(middle_east_df, "gross_inc_percap")

print("Middle East HDI Outliers:")
print(me_hdi_outliers[["country", "hdi"]])

print("\nMiddle East GNI per Capita Outliers:")
print(me_gni_outliers[["country", "gross_inc_percap"]])
```

```python
# Scatter Plot: South Asia (HDI vs GNI per Capita)
plt.figure(figsize=(8,6))

# Normal countries
plt.scatter(
south_asia_df["gross_inc_percap"],
    south_asia_df["hdi"],
    color='blue',
    label="Normal Countries"
)

# Outliers
plt.scatter(
    sa_gni_outliers["gross_inc_percap"],
    sa_hdi_outliers["hdi"],
    color='red',
    label="Outliers"
)
```

```python
# scatter Plot: Middle East (HDI vs GNI per Capita)
plt.figure(figsize=(8,6))

# Normal countries
plt.scatter(middle_east_df["gross_inc_percap"], middle_east_df["hdi"], color="blue", label="Normal")

# Combine HDI and GNI outliers for Middle East
me_outliers = pd.concat([me_hdi_outliers, me_gni_outliers]).drop_duplicates()

# Outliers
plt.scatter(me_outliers["gross_inc_percap"], me_outliers["hdi"], color="red", label="Outliers")

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("Middle East: HDI vs GNI per Capita (Outliers Highlighted)")
plt.legend()
plt.show()
```

# Discussion: Significance of Outliers

- Outliers identify countries whose HDI or GNI per Capita deviate significantly from the regional pattern.
- **High GNI but low HDI**: Indicates income is not translating into human development (e.g., inequality or poor social services).
- **Low GNI and low HDI**: Suggests structural challenges such as poverty, conflict, or poor access to education and healthcare.
- Analyzing outliers is crucial for **targeted policy interventions**, as they highlight countries that need more attention despite high income or reveal hidden developmental weaknesses.
