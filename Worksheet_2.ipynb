Problem - 4:

Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a
visual ization to show how the survival distributions vary by gender.

```python
mean_survival = data.groupby('Sex')['Survived'].mean()
print(mean_survival)
```

```python
import matplotlib.pyplot as plt

mean_survival.plot(kind='bar', color=['pink','blue'])

plt.xlabel('Gender')
plt.ylabel('Mean Survival Rate')
plt.title('Mean Survival Rate by Gender')
plt.ylim(0, 1)

for i,v in enumerate(mean_survival):
  plt.text(i, v + 0.02, f"{v:.2f}", ha ='center')

plt.show()
```

Problem - 5:

Draw a visualization that breaks your visualization from Exercise 3 down by the port of embarkation
(’Em barked’). In this instance, compare the ports ’C’ (Cherbourg), ’Q’ (Queenstown), and ’S’
(Southampton).

```python
data = pd.read_csv("/content/drive/MyDrive/Concepts and technology of AI/Titanic-Dataset.csv")
```

```python
survival_by_port_gender = data.groupby(['Embarked', 'Sex'])['Survived'].mean().unstack()
print(survival_by_port_gender)
```

```python
survival_by_port_gender.plot(kind='bar', color=['pink', 'blue'])

plt.title('Mean Survival Rate by Gender and Port of Embarkation')
plt.ylabel('Mean Survival Rate')
plt.xlabel('Port of Embarkation')
plt.ylim(0,1)

for i, port in enumerate(survival_by_port_gender.index):
  for j, gender in enumerate(survival_by_port_gender.columns):
    value = survival_by_port_gender.loc[port, gender]
    plt.text(i + j*0.25 - 0.125, value + 0.02, f"{value:.2f}", ha='center', va='center')

plt.legend(title='Gender')
plt.show()
```

Problem - 6{Optional}:

Show how the survival rates (’Survived’) vary by age group and passenger class (’Pclass’). Break up
the ’Age’ column into five quantiles in your DataFrame, and then compare the means of ’Survived’ by
class and age group. Draw a visualization using a any plotting library to represent this graphically.

```python
data = data.dropna(subset=['Age', 'Pclass'])
```

```python
data['AgeGroup'] = pd.qcut(data['Age'],5)
print(data[['Age','AgeGroup']].head())
```

```python
survival_by_age_pclass = data.groupby(['AgeGroup', 'Pclass'])['Survived'].mean().unstack()
print(survival_by_age_pclass)
```

```python
survival_by_age_pclass.plot(kind='bar', figsize=(10,6), width=0.8)

plt.title('Mean Survival Rate by Age Group and Passenger Class')
plt.xlabel('Age Group')
plt.ylabel('Mean Survival Rate')
plt.ylim(0,1)
plt.xticks(rotation=45)
for i, age_group in enumerate(survival_by_age_pclass.index):
  for j, pclass in enumerate(survival_by_age_pclass.columns):
    value = survival_by_age_pclass.loc[age_group, pclass]
    plt.text(i + j*0.2 - 0.2, value + 0.02, f"{value:.2f}", ha='center', va='center', fontsize=8)

plt.legend(title='Passenger Class')
plt.show()
```
