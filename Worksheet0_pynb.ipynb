```python
time_data = [
(3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
(4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
(5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
(3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
(4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]

#Creating empty lists for study time classifications
low_study_time = [] # For study time lesser than 3 hours
moderate_study_time = [] # For study hours between 3 and 5 hours
high_study_time = [] # For study hours more than 5 hours

#Iterate and classify study hour
for times in time_data:
    if times[0] < 3:
        low_study_time.append(times)
    elif times[0] >= 3 and times[0] <=5:
         moderate_study_time.append(times)
    else:
         high_study_time.append(times)


#Print lists to verify classifications
print("Low study time list : {low_study_time}")
print("Moderate study time list : {moderate_study_time}")
print("High study time list : {high_study_time}")
```

```python
#Task 1 - Sum of Nested Lists
def sum_nested_list(nested_list):
    total = 0
    for item in nested_list:
        if isinstance(item, list):
            total += sum_nested_list(item)
        else:
            total += item
    return total

nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
total_sum = sum_nested_list(nested_list)
print(f"The sum of numbers in the list is : {total_sum}")
```

```python
#Task 2 - Generate All Permutation of a String

def generate_permutations(s):

    if len(s) == 1:
        return [s]
    total_characters = []
    for character in s:
        if character in total_characters:
            continue
        else:
            total_characters.append(character)

    new_character = ""
    for item in total_characters:
        new_character += item
    output = []
    for i, ch in enumerate(new_character):
        output += [ch + p for p in generate_permutations(new_character[:i] + new_character[i+1:])]
    return output

ch = "aaabc"
print(generate_permutations(ch))
```

```python
#Task 3 - Directory Size Calculation

def calculate_directory_size(directory):
    total_size = 0
    for k,v in directory.items():
        if isinstance(v, dict):
            total_size += calculate_directory_size(v)
        else:
            total_size += v
    return total_size

directory_structure = {
"file1.txt": 200,
"file2.txt": 300,
"subdir1": {
"file3.txt": 400,
"file4.txt": 100
},
"subdir2": {
"subsubdir1": {
"file5.txt": 250
},
"file6.txt": 150
}
}

total_directory_size = calculate_directory_size(directory_structure)
print(f"Total directory size : {total_directory_size} KB")
```

```python
#1

def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

print(min_coins([1, 2, 5], 11))
```

```python
#2
def longest_common_subsequence(s1, s2):
    """
    Uses dynamic programming to find the length of the longest common subsequence between two strings.
    """
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

s1 = "abcde"
s2 = "ace"
print("Length of LCS:", longest_common_subsequence(s1, s2))
```

```python
#3
def knapsack(weights, values, capacity):
    """ Uses dynamic programming to solve the 0/1 Knapsack Problem.
    Returns the maximum value that can be achieved within the given capacity.
    """
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print("Maximum value:", knapsack(weights, values, capacity))
```
