**Prove: A.A-1=I**

```python
A = np.array([[3, 4], [7, 8]])

B = np.array([[5, 3],[2, 1]])

A_inv = np.linalg.inv(A)
I = A @ A_inv
print(I)
```

**AB/=BA**

```python
AB = A @ B
BA = B @ A
print("AB = \n", AB)
print("BA = \n", BA)
```

**(AB)^T = B^T.A^T**

```python
left = (A @ B).T
right = B.T @ A.T

print("Left = \n", left)
print("Right = \n", right)
```

**Solve System of Linear Equations using Inverse Method**

2x-3y+z = -1x-y+2z = -3 3x+y-z = 9

```python
A = np.array([[2,-3,1],
              [1,-1,2],
              [3,1,-1]])

B = np.array([-1,-3,9])
A_inv = np.linalg.inv(A)
X = A_inv @ B
print(X)
```
